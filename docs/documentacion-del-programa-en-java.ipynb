# Clase lechuga:

## Declaración del paquete:


```python
package classes;
```

Este código se encuentra en un paquete llamado "classes".

## Declaración de la clase Lechuga:

```python
public class Lechuga {
```

Aquí se declara la clase Lechuga. La palabra clave public significa que esta clase es accesible desde cualquier otra clase. El nombre de la clase es Lechuga.


## Atributos de la clase:

```python
private int id;
private String fechaPlantacion;
private double temperatura;
private double humedad;
```

Estos son los atributos de la clase Lechuga. Son privados (private), lo que significa que solo pueden ser accedidos directamente dentro de la clase. Cada lechuga tiene un id (número entero), una fechaPlantacion (cadena de caracteres), una temperatura (número decimal de punto flotante) y una humedad (también número decimal de punto flotante).

## Constructor:

```python
public Lechuga(int id, String fechaPlantacion, double temperatura, double humedad) {
    this.id = id;
    this.fechaPlantacion = fechaPlantacion;
    this.temperatura = temperatura;
    this.humedad = humedad;
}
```

Este es el constructor de la clase Lechuga. Se utiliza para crear objetos de la clase Lechuga. Toma cuatro parámetros: id, fechaPlantacion, temperatura y humedad, y los asigna a los atributos correspondientes de la instancia de la clase (this.id, this.fechaPlantacion, this.temperatura y this.humedad). El this se utiliza para referirse a los atributos de la instancia actual de la clase.

## Método getInfo():

```python
public String getInfo() {
    return "ID: " + id + ", Fecha de Plantación: " + fechaPlantacion + ", Temperatura: " + temperatura + ", Humedad: " + humedad;
}
```

Este método devuelve la información de la lechuga en forma de cadena de caracteres. Concatena el id, fechaPlantacion, temperatura y humedad en una cadena de texto que se devuelve.

## Método getId():

```python
public int getId() {
    return id;
}
```

Este método devuelve el id de la lechuga.

# Clase RegistroLechugas:

## Declaración del paquete:

```python
package classes;
```

Este código se encuentra en el paquete "classes", al igual que la clase Lechuga. 

## Importaciones:



```python
import java.io.FileWriter;
import java.io.IOException;
import java.io.PrintWriter;
import java.util.ArrayList;
```

Se importan algunas clases necesarias para el funcionamiento de esta clase. FileWriter y PrintWriter se utilizan para escribir en archivos de texto, IOException se utiliza para manejar excepciones de entrada/salida, y ArrayList se utiliza para almacenar la lista de lechugas.

## Declaración de la clase RegistroLechugas:

```python
public class RegistroLechugas {
```

Aquí se declara la clase RegistroLechugas. La palabra clave public significa que esta clase es accesible desde cualquier otra clase. El nombre de la clase es RegistroLechugas.

## Atributos de la clase:

```python
private ArrayList<Lechuga> lechugas;
```

Este es el único atributo de la clase RegistroLechugas. Es una lista de objetos de tipo Lechuga. Se utiliza para almacenar las lechugas registradas.

## Constructor:

```python
public RegistroLechugas() {
    lechugas = new ArrayList<>();
}
```

Este es el constructor de la clase RegistroLechugas. Se utiliza para inicializar la lista lechugas como una nueva instancia de ArrayList. Esto significa que cuando se crea un nuevo objeto RegistroLechugas, la lista de lechugas está vacía.

## Método agregarLechuga(Lechuga lechuga):

```python
public void agregarLechuga(Lechuga lechuga) {
    lechugas.add(lechuga);
}
```

Este método se utiliza para agregar una lechuga al registro. Toma un objeto Lechuga como parámetro y lo agrega a la lista lechugas.

## Método guardarRegistro():

```python
public void guardarRegistro() {
    try (PrintWriter writer = new PrintWriter(new FileWriter("registro_lechugas.txt"))) {
        for (Lechuga lechuga : lechugas) {
            writer.println(lechuga.getInfo());
        }
    } catch (IOException e) {
        e.printStackTrace();
    }
}
```

Este método se utiliza para guardar el registro de lechugas en un archivo de texto llamado "registro_lechugas.txt". Utiliza un bloque try-with-resources para abrir un PrintWriter que escribirá en un FileWriter. Luego, itera sobre cada lechuga en la lista lechugas y escribe la información de cada lechuga en una línea del archivo. En caso de que ocurra un error de entrada/salida durante la escritura del archivo, se maneja la excepción e imprime el rastreo de la pila del error.
